```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Sales Data.csv")
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.isna().sum() #shows that there are no null values
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)}({get_state(x)})")
df.head()
```

What was the best month for sales? How much was earned that month?

```python
results = df.groupby('Month').sum()
```

```python
months=range(1,13)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.xlabel("Month number")
plt.ylabel("Sales in USD ($)")
plt.show()
```

What city sold the most product?

```python
city_sales=df.groupby('City')['Sales'].sum().nlargest(5)
city_sales
```

```python
sns.barplot(y=city_sales.index,x=city_sales).set(title='Top 5 cities which sold the most' );
```

What time should we display advertisemens to maximize the likelihood of customers buying product?

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Add hour column
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
df['Minute'] = pd.to_datetime(df['Order Date']).dt.minute
df['Count'] = 1
df.head()
```

```python
keys = [pair for pair, df in df.groupby(['Hour'])]

plt.plot(keys, df.groupby(['Hour']).count()['Count'])
plt.xticks(keys)
plt.grid()
plt.xlabel('Hours')
plt.ylabel('Number of Orders')
plt.show()
```

What products are most often sold together?


```python
df1 = df[df['Order Date'].duplicated(keep=False)]

df1['Grouped'] = df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df1 = df1[['Order ID', 'Grouped']].drop_duplicates()
df1.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df1['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

What product sold the most? Why do you think it sold the most?

```python
product_group = df.groupby('Product')
```

```python
quantity_ordered = product_group.sum(numeric_only=True)['Quantity Ordered']

products = [product for product ,df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products,rotation='vertical',size=8)
plt.show()
```

```python
# Convert 'Price Each' column to numeric type
df['Price Each'] = pd.to_numeric(df['Price Each'])

# Calculate prices
prices = df.groupby('Product')['Price Each'].mean() # Select the 'Price Each' column after groupby

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products , prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
